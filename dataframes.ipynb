```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
df
```

```python
# a. Create a column named passing_english that indicates whether 
# each student has a passing grade in english.

new_df = df.assign(passing_english = df.english >= 70)
new_df
```

```python
# b. Sort the english grades by the passing_english column. 
# How are duplicates handled?

new_df.sort_values(by = 'passing_english')


# they are sorted by passing engish or not. The grades are not in any order.
```

```python
# c. Sort the english grades first by passing_english and then by 
# student name. All the students that are failing english should 
# be first, and within the students that are failing english they 
# should be ordered alphabetically. The same should be true for the 
# students passing english. (Hint: you can pass a list to the 
# .sort_values method)

new_df.sort_values(by=['passing_english', 'name'], ascending = [True, True])  
```

```python
# d. Sort the english grades first by passing_english, and then by 
#the actual english grade, similar to how we did in the last step.

new_df.sort_values(by=['passing_english', 'english'], ascending = [True, True])  
```

```python
# e. Calculate each students overall grade and add it as a column on 
# the dataframe. The overall grade is the average of the math, 
# english, and reading grades.

new_df = df.assign(overall_grade = (df.math+df.english+df.reading)/3)
new_df.round(0)
```

```python
# d. Rename the cty column to city.
new_mpg = mpg.rename(columns = {'cty':'city'})
new_mpg
```

```python
# e. Rename the hwy column to highway.
new_mpg = new_mpg.rename(columns = {'hwy':'highway'})
new_mpg
```

```python
# f. Do any cars have better city mileage than highway mileage?
better_city_mpg = new_mpg.sort_values(by=['city', 'highway'], ascending = [False, False])
better_city_mpg
```

```python
mask = new_mpg = new_mpg.city > new_mpg.highway
len(mpg[mask])
```

```python
(mpg.cty > mpg.hwy).all()
```

```python
# g. Create a column named mileage_difference this column should 
# contain the difference between highway and city mileage for each 
# car.
mil_diff = (new_mpg.assign(mileage_difference = (new_mpg.highway - new_mpg.city))
.sort_values(by = 'mileage_difference', ascending = False))
mil_diff
```

```python
# h. Which car (or cars) has the highest mileage difference?
mil_diff.nlargest(1, 'mileage_difference', keep='all')
```

```python
# i. Which compact class car has the lowest highway mileage? The best?

new_mpg_low = (new_mpg[new_mpg['class'] == 'compact'].sort_values(by = 'highway', ascending = False).tail(1))
new_mpg_low
```

```python
new_mpg_high = (new_mpg[new_mpg['class'] == 'compact']
.sort_values(by = 'highway', ascending = False)
.head(1))
new_mpg_high
```

```python
# j. Create a column named average_mileage that is the mean of the 
# city and highway mileage.

avg_mpg = (new_mpg.assign(average_mileage = (new_mpg.highway + new_mpg.city)/2)
.sort_values(by = 'average_mileage', ascending = False))
avg_mpg
```

```python
# k. Which dodge car has the best average mileage? The worst?

new_mpg_h = (avg_mpg[new_mpg['manufacturer'] == 'dodge']
.sort_values(by = 'average_mileage', ascending = False)
.head(1))
new_mpg_h
```

```python
new_mpg_l = (avg_mpg[new_mpg['manufacturer'] == 'dodge']
.sort_values(by = 'average_mileage', ascending = False)
.tail(1))
new_mpg_l
```
