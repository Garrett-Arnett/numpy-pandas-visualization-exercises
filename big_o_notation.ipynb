```python
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint
import random
import pandas as pd


'''5. Make a new Jupyter notebook named big_o_notation.ipynb

Title your chart "Big O Notation"
Label your x axis "Elements"
Label your y axis "Operations"
Label your curves or make a legend for the curves
Use LaTex notation where possible
Curves to graph:'''
```

```python

x = range(1, 30)

y1 = [x for x in x]
y2 = [math.log(x) for x in x]
y3 = [x for x in x]
y4 = [x*math.log(x) for x in x]
y5 = [x**2 for x in x]
y6 = [2**x for x in x]
y7 = [math.factorial(x) for x in x]
y8 = [x**x for x in x]

plt.figure(figsize = (10,6))
plt.title('Big O Notation', color = 'w', weight = 'bold')
plt.xlabel = ('Elements')
plt.ylabel = ('Operations')
plt.style.use('seaborn-dark-palette')

plt.plot(x, y1,linestyle = '--', marker = 'o', color = 'k', label = 'O(1)')
plt.plot(x, y2,linestyle = '--', marker = '+', color = 'b', label = 'O(log n)')
plt.plot(x, y3,linestyle = '--', marker = 'v', color = 'm', label = 'O(n)')
plt.plot(x, y4,linestyle = '--', marker = '.', color = 'r', label = 'O(n log n)')
plt.plot(x, y5,linestyle = '--', marker = 'o', color = 'r', label = 'O(n^2)')
plt.plot(x, y6,linestyle = '--', marker = '|', color = 'g', label = 'O(2^n)')
plt.plot(x, y7,linestyle = '--', marker = '<', color = 'y', label = 'O(n!)')
plt.plot(x, y8,linestyle = '--', marker = '>', color = 'k', label = 'O(n^n)')

plt.xlim(1,11)
plt.ylim(0,50)

plt.legend()
plt.show()
```
