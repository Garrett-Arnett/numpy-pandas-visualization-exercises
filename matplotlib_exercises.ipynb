```python
# 1. Use matplotlib to plot the following equation.  y=x^2 - x + 2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.
plt.style.use('seaborn-dark-palette')
plt.figure(figsize = (10,6))
x = range(-50, 50, 2)
y = [ n ** 2 - n for n in x]
plt.scatter(x,y)
plt.title('$y = x^2-x+2$', color = 'w')

plt.xlabel('$x$', fontsize = 14, color = 'w')
plt.ylabel('$x^2-x+2$', fontsize = 14, color = 'w')
plt.grid(True, ls = '--')

plt.annotate('The Origin', xy = (0, 0), xytext = (0, 500), arrowprops = dict(facecolor = 'yellow'))
```
